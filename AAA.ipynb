```python
# 1. Line Plot: Close price over time
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Close'], label='Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Close Price Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['Date'], data['Volume'], label='Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Volume Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(9, 6))
plt.scatter(data['High'], data['Low'], alpha=0.5, label='High vs. Low')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.title('High vs. Low Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 6. Open vs. Close Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Open'], data['Close'], alpha=0.5, label='Open vs. Close')
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.title('Open vs. Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 7. Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
mask = np.triu(np.ones_like(data.corr(), dtype=bool))
sns.heatmap(data.corr(), annot=True, mask=mask, fmt='.2f', vmin=0)
```

```python
# Perform label encoding on the specified columns and display the unique values for each encoded column. 
# Finally, it will print the updated DataFram

def LABEL_ENCODING(c1):
    from sklearn import preprocessing
    label_encoder = preprocessing.LabelEncoder()
    data[c1]= label_encoder.fit_transform(data[c1])
    data[c1].unique()
LABEL_ENCODING("High")
LABEL_ENCODING("Low")
LABEL_ENCODING("Close")
data
```
